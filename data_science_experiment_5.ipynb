Load the Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from UCI ML Repository
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, delimiter=';')

# Display first few rows
print(df.head())
```

Data Visualizations

```python
plt.figure(figsize=(8, 5))
plt.hist(df['quality'], bins=7, color='skyblue', edgecolor='black')
plt.title("Distribution of Wine Quality Ratings")
plt.xlabel("Quality")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Scatter Plot

```python
plt.figure(figsize=(8, 5))
plt.scatter(df['alcohol'], df['quality'], color='red', alpha=0.5)
plt.title("Alcohol Content vs Wine Quality")
plt.xlabel("Alcohol Content (%)")
plt.ylabel("Quality Rating")
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```

Boxplot

```python
plt.figure(figsize=(8, 5))
plt.boxplot(df['pH'], vert=False, patch_artist=True, boxprops=dict(facecolor="lightgreen"))
plt.title("Boxplot of pH Levels in Wine")
plt.xlabel("pH Level")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, delimiter=';')

quality_counts = df['quality'].value_counts()

plt.figure(figsize=(8, 6))
plt.pie(quality_counts, labels=quality_counts.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99','#ffcc99','#c2c2f0','#ffb3e6'], startangle=140)
plt.title("Distribution of Wine Quality Scores")
plt.axis('equal')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

url = "https://archive.ics.uci.edu/ml/machine-learning-databases/wine-quality/winequality-red.csv"
df = pd.read_csv(url, delimiter=';')

quality_mean = df.groupby('quality').mean()

plt.figure(figsize=(10, 5))
for feature in ['alcohol', 'volatile acidity', 'citric acid', 'sulphates']:
    plt.plot(quality_mean.index, quality_mean[feature], marker='o', label=feature)

plt.xlabel("Wine Quality Score")
plt.ylabel("Average Feature Value")
plt.title("Wine Quality vs. Key Features")
plt.legend()
plt.grid(True)
plt.show()
```
